Task-02Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
titanic = pd.read_csv('/content/titanic.csv')
# Check for missing values
print(titanic.isnull().sum())

# Deal with missing values, for example:
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
# Check and convert data types if necessary
titanic['Age'] = titanic['Age'].astype(int)
# Summary statistics
print(titanic.describe())
# Explore the distribution of passengers' ages
sns.histplot(titanic['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Relationship between passenger class and survival
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival by Passenger Class')
plt.show()

# Relationship between gender and survival
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Survival by Gender')
plt.show()
# Create a correlation matrix
correlation_matrix = titanic.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
